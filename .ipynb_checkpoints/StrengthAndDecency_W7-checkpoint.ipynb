## Strength and decency, Roosevelt

##### From my Natural Language Processing analysis of this poem, I think that several conclusions can be drawn. Firstly, I think that Roosevelt tries to strengthen the morale of his audience. The intended audience seems to be male, defined by the frequency of words such as "man", "men", "boy". These groups, as the concordance view shows, are often addressed as "decent", "honorable", "strong, "good"; emphasising their positive example for society.
##### What I found interesting is the distribution view of the text. I have combined the nouns and adjecives and have found that Roosevelt seems to talk to a group, and emphasise a communal effort, as he uses the word "men" frequently at the beginning and end, often in combination with some of adjectives mentioned above. The words "boys" and "boy" however occur in the middle of the text. It seems that the text follow a development of the audience, now decent adults, who had to overcome the foolishness of boyhood. On the other side (at the end of the poem), they emerge as men, who have found the decency and strength that he proclaims.

```python
import nltk
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
poem = open('StrengthAndDecency_TRoosevelt.txt', 'r', encoding = 'utf-8')
Strength = poem.read()
Strength
```

```python
StrengthLower = Strength.lower()
StrengthLower
```

```python
StrengthTok = nltk.word_tokenize(StrengthLower)
print(StrengthTok)
```

```python
from nltk.stem import WordNetLemmatizer
lemm = WordNetLemmatizer()
```

```python
StrengthLemm = []
for w in StrengthTok:
    x = lemm.lemmatize(w, pos='v')
    StrengthLemm.append(x)
print(StrengthLemm)
```

#### The lemmatised list still had various verb forms, to I tried to eliminate those.

```python
from nltk.corpus import stopwords
Stopwords = set(stopwords.words('english'))
Stopwords
```

```python
StrengthClean = [w for w in StrengthLemm if w not in Stopwords]
StrengthClean
```

```python
Strengthpunc = ['.',',',':','“','”','–','’','"',"'"]
Strengthpunc
```

```python
StrengthClean2 = [w for w in StrengthClean if w not in Strengthpunc]
StrengthClean2
```

## Frequency

```python
from nltk.probability import FreqDist
StrengthFreq = FreqDist(StrengthClean2)
StrengthFreq
```

## Vizualisation

```python
%matplotlib inline
plt.figure(figsize = (30,10))
plt.xticks(fontsize = 20)
plt.title('Strength and Decency, Frequency')
StrengthFreq.plot(40)
```

```python
Cleaner2 = [';','u','wa','ha','!','?']
StrengthClean3 = [w for w in StrengthClean2 if w not in Cleaner2]
StrengthClean3
```

```python
%matplotlib inline
StrengthFreq2 = FreqDist(StrengthClean3)
plt.figure(figsize=(30,10))
plt.xticks(fontsize=20)
plt.title('Strength and Decency Clean, Frequency')
StrengthFreq2.plot(50)
```

```python
StrengthCommon = StrengthFreq2.most_common(20)
StrengthCommon
```

```python
df = pd.DataFrame(StrengthCommon, columns = ['Words','Counts'])
df
```

```python
df.plot(kind='bar', x='Words', y='Counts', legend = True, color=('red','blue'))
plt.xlabel('Words')
plt.ylabel('Counts')
plt.title('Strength and Decency, Cleaner Frequency')
```

```python
df.plot(kind='line', x='Words', y='Counts', legend = True, color=('green'))
plt.xticks(fontsize=6)
plt.xlabel('Words')
plt.ylabel('Counts')
plt.title('Strength and Decency, Cleaner Frequency')
```

```python
from nltk.text import Text
```

```python
StrengthObj = nltk.Text(StrengthClean3)
StrengthObj
```

```python
StrengthObj.concordance("man")
StrengthObj.concordance("men")
StrengthObj.concordance("young")
StrengthObj.concordance("life")
StrengthObj.concordance("strong")
```

#### Highly frequent words tend to occur together, potential to reiterate Roosevelt's message to his audience. The word 'man' is used in association with positive associations ('clean mouth, good citizen, able hold rough work, unselfish').

```python
StrengthObj.dispersion_plot(['man','life','strong','boy','boys','foul','men','young','good','work','decent'])
```

#### The dispersion across the text is in line with the concordance view of 'boy' and 'young', emphasising the correlation. While Roosevelt tals baout boy, he does not seem to talk about adult aged men. Also, Roosevelt seems to speak more often about men as a group rather than the indivdual ('man'), possibly highlighting the importance of communal values.
